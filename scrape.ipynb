## Craping the Data from "Find Easy" website

```python
import os
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the website
url = "https://www.findeasy.in/tobacco-use-among-adults-in-india/"

# Send a request to the website and get the HTML content
response = requests.get(url)
html_content = response.text

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(html_content, "html.parser")

# Find the table body (tbody) element
table_body = soup.find("tbody")

# Create a list to store the data rows
data_rows = []

# Find all the rows (tr) within the table body
for idx, row in enumerate(table_body.find_all("tr")):
    # Skip the first row (contains column headers)
    if idx == 0:
        continue
    # Find all the cells (td) within each row
    cells = row.find_all("td")
    # Extract the data and store it in a list
    row_data = [cell.text.strip() for cell in cells]
    # Append the row data to the list of data rows
    data_rows.append(row_data)

# Create a DataFrame from the list of data rows
df = pd.DataFrame(data_rows, columns=["Sl. No", "States & Union Territories", "Women (15 years & above)", "Men (15 years & above)"])

```
